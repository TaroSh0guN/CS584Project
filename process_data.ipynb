```python
import pandas as pd
import re
import nltk
import ssl
from sklearn.model_selection import train_test_split

try:
    _create_unverified_https_context = ssl._create_unverified_context
except AttributeError:
    pass
else:
    ssl._create_default_https_context = _create_unverified_https_context

nltk.download('stopwords')
nltk.download('punkt')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('english'))

snap_df = pd.read_json("unfiltered_data/news-snapchat-noagegroup-mental-2024.jsonl", lines=True)
yt_df = pd.read_json("unfiltered_data/news-youtube-noagegroup-mental-2024.jsonl", lines=True)

def split(df, source):
    df = df.dropna(subset=['body'])
    df['body'] = df['body'].apply(lambda x: ' '.join(x) if isinstance(x, list) else str(x))

    def tokenize(text):
        text = text.lower()
        text = re.sub(r'[^a-z\s]', '', text)
        tokens = word_tokenize(text)
        tokens = [word for word in tokens if word not in stop_words]
        return tokens

    df['tokens'] = df['body'].apply(tokenize)

    train_df, temp_df = train_test_split(df, test_size=0.3)
    val_df, test_df = train_test_split(temp_df, test_size=0.5)

    print(train_df.head())
    print(train_df.head())
    print(train_df.head())

    #train_df.to_json(f"train_{source}_articles.jsonl", orient='records', lines=True, force_ascii=False)
    #val_df.to_json(f"val_{source}__articles.jsonl", orient='records', lines=True, force_ascii=False)
    #test_df.to_json(f"test_{source}_articles.jsonl", orient='records', lines=True, force_ascii=False)

split(snap_df, "snap")
split(yt_df, "yt")
```
